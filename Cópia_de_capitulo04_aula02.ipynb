### Estrutura de Repetição `while`

image.png

```python
n = 0
while n <= 15 and n == 0:
    print(n, 'Meu nome é Antônio')
    n = n + 1
print('Fim do programa')
```

```python
#Escrevendo meu nome 15 vezes
n = 15 #variável controle
while n > 0:
    print(n, 'Meu nome é André Luiz')
    n = n - 1 #diminuindo o valor de 15 até chegar em zero
print('Fim do programa')
```

```python
#Escrevendo meu nome 15 vezes
n = 1 #variável controle
while n <= 15:
    print(n, 'Meu nome é André Luiz')
    n = n + 1 #diminuindo o valor de 15 até chegar em zero
print('Fim do programa')
```

```python
#Escrevendo meu nome 15 vezes
n = 1 #variável controle
while n <= 15 and n != 0:
    print(n, 'Meu nome é André Luiz')
    n = n + 1 #diminuindo o valor de 15 até chegar em zero
print('Fim do programa')
```

```python
#Escrevendo meu nome 15 vezes
n = 0 #variável controle
while n <= 15 and n == 0:
    print(n, 'Meu nome é André Luiz')
    n = n + 1 #diminuindo o valor de 15 até chegar em zero
print('Fim do programa')
```

```python
#Escrevendo meu nome 15 vezes
n = 1 #variável controle
while n <= 15 and n == 1:
    print(n, 'Meu nome é André Luiz')
    n = n + 1 #diminuindo o valor de 15 até chegar em zero
print('Fim do programa')
```

```python
# exemplo com n = 15
n = 15
soma = 0
contador = 0

while contador <= n:
    # print('soma', soma)
    # print('contador', contador)
    soma = soma + contador
    contador = contador + 1
    # print('soma depois da execução', soma)
    # print('contador depois da execução', contador)
    # print()

print(f'A soma dos primeiros {n} inteiros é {soma}')
```

```python
# exemplo com n = 15 de soma dos n primeiros inteiros
n = 15
soma = 0
contador = 0

while contador <= n:
  print('soma', soma)
  print('contador', contador)

  soma = soma + contador
  contador = contador + 1
  print('soma depois da execução', soma)
  print('contador depois da execução', contador)
  #print()

print(f'A soma dos primeiros {n} inteiros é {soma}')
```

```python
# exemplo com n = 15
n = 15
soma = 0

while n >= 0:
    soma = soma + n
    n = n - 1

print(f'A soma dos primeiros inteiros é {soma}')
```

```python
#exemplo de palavra araraquara com print no escopo for
p = 'Araraquara'
contador = 0
for texto in p:
  if texto == 'a':
    contador = contador + 1
print(f'A palavra {p} possui {contador} letras "a".') #influência da identação
print('FIM')
```

```python
texto = input('Digite um nome: ')

vogais = 'AEIOU'
print(f'A sequência de vogal/ vogais na palavra {texto}: ')
for c in texto:
  if c.upper() in vogais:
    print(c, end='') 
```

```python
texto = input('Digite um nome: ')

vogais = 'AEIOU'
print(f'A sequência de vogal/ vogais na palavra {texto}: ')
for c in texto:
  if c.upper() in vogais:
    print(c + '|', end='') #com um separador entre as vogais
```

```python
# exemplo de palavra: araraquara
p = 'araraquara'
contador_A = 0
contador_R = 0

for letra in p:
    if letra == 'a':
        contador_A = contador_A + 1
    if letra == 'r':
        contador_R = contador_R + 1
print(f"A palavara {p} possuí {contador_A} letras 'a' e {contador_R} letras 'r'")
```

### Interrupção da Estrutura de Repetição

```python
while True:
    print('Antonio Carlos')
    break
```

```python
# for <item> in <sequência>:
#     <bloco de código a ser repetido>
#     if <expressao>
#         <outro bloco de código>
#         break
```

```python
# Realiza a busca na faixa de 250 e 300
for num in range(250, 301):
    # verifica se o número é divisível por 21
    if num % 21 == 0:        
        print('Número encontrado:', num)
        # se for divísivel por 21, interrompe a busca
        break
```
